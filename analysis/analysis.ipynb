```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the csv called log_2023-09-27 20:07:30.242808.csv which is in the game_logs folder in the parent directory
data_path = "../game_logs/"
df = pd.read_csv(data_path+'log_2023-09-27 21:18:36.265519.csv')
```

```python
#visualize the river_dirtiness, orchard_fruits, and agent_reward over time. The reward should be plotted on a different scale than the other two
fig, ax1 = plt.subplots(figsize=(12, 8))
ax2 = ax1.twinx()
ax1.plot([i for i in range(len(df['river_dirtiness']))], df['river_dirtiness'], label='river dirtiness')
ax1.plot([i for i in range(len(df['orchard_fruits']))], df['orchard_fruits'], label='orchard fruits')
ax2.plot([i for i in range(len(df['agent_reward']))], df['agent_reward'], label='agent reward', color='red')
ax1.set_xlabel('time')
ax1.set_ylabel('river dirtiness and orchard fruits')
ax2.set_ylabel('agent reward')
fig.legend()
plt.show()
```
